## Create range of dates with pd.date_range Function

- la función date_range genera y retona un DatetimeIndex que contiene una secuencia de fechas
- la función necesita al menos 2 de 3 parámetros: start, end y period
- conocidos start y end , pandas asumirá un perido/intervalo diario
- cada elemento de un DatetimeIndex es un Timestamp

```python
pd.date_range(start="2025-01-01", end="2025-01-07")
# es lo mismo que:
pd.date_range(start="2025-01-01", end="2025-01-07", freq="D")

#intervalo  de 2 días
pd.date_range(start="2025-01-01", end="2025-01-07", freq="2D")

# intervalo business days Monday to Friday 
pd.date_range(start="2025-01-01", end="2025-01-07", freq="B") # se excluyen días 4 y 5

# intervalo semanal "W" ( por defecto domingo - EEUU)       
pd.date_range(start="2025-01-01", end="2025-01-31", freq="W")

# intervalo semanal comenzando los viernes
pd.date_range(start="2025-01-01", end="2025-01-31", freq="W-FRI")

# intervalo semanal comenzando los jueves
pd.date_range(start="2025-01-01", end="2025-01-31", freq="W-THU")

# por horas
pd.date_range(start="2025-01-01", end="2025-01-31", freq="H")
pd.date_range(start="2025-01-01", end="2025-01-31", freq="6H")

# por fin de mes
pd.date_range(start="2025-01-01", end="2025-12-31", freq="M")
# por inicio de mes
pd.date_range(start="2025-01-01", end="2025-12-31", freq="MS")

# por fin de año 
pd.date_range(start="2025-01-01", end="2050-12-31", freq="A")
```

```python
# forma alternativa de utilizar esta función

pd.date_range(start= "2012-09-09",freq="D", periods=25 )
pd.date_range(start= "2012-09-09",freq="3D", periods=40 )

pd.date_range(start= "2012-09-09",freq="B", periods=180 )
```

```python
# de la misma forma se puede entregar la fecha final 

pd.date_range(end= "2013-10-31",freq="D", periods=20 )

pd.date_range(end= "2013-12-31",freq="B", periods=75)
```

## Selecting rows from a DF with a DatetimeIndex
- el accesor Iloc está disponible para extraccción basada en posición  
- el accesor loc acepta strings ó timestamps para extraer por etiquetas de  índices (valores). mientras que los obejtos datetime no funcionaran
- se recomienda utilizar slicing para obtener una secuencia de valores, un alternativa es el método truncate

```python
# importamos ibm.csv
pd.read_csv("ibm.csv", parse_dates=["Date"])
# exploramos sus dtypes
pd.read_csv("ibm.csv", parse_dates=["Date"]).dtypes
# al observar que la columna Date es un objeto datetime la usaremos como index
stocks =pd.read_csv("ibm.csv", parse_dates=["Date"], index_col="Date").sort_index()
stocks.head()
```

```python
# utilizamos los accesor
stocks.iloc[300]
stocks.loc["2014-03-04"]
# utilizamos panda para acceder a Timestamps
stocks.loc[pd.Timestamp(2014,3,4)]
#slicing
stocks.loc["2014-03-04":"2014-12-31"]
# del mismo modo podemos utilizar Timestaps de pandas
stocks.loc[pd.Timestamp(2014,3,4): pd.Timestamp(2014,12,31)]

# utilizando truncate, parametros before y after
stocks.truncate("2014-03-04","2014-12-31")

#con loc podemos filtrar etiquetas o headers
stocks.loc["2014-03-04", "Close"]
stocks.loc["2014-03-04", "High":"Close"]

#  ahora con timestamps
stocks.loc[ pd.Timestamp(2014,3,4): pd.Timestamp(2014,12,31), "High":"Close"]
```

## The DateOffset object
- el objeto DateOffset agrega tiempo a un Timestamp para crear un nuevo Timestamp
- el constructor DatteOffset acepta parametros como: days, weeks, months, years y otros
- podemos entregar un objeto DateOffset al parámetro freq de una función pd.date_range

```python
stocks =pd.read_csv("ibm.csv", parse_dates=["Date"], index_col="Date").sort_index()
stocks.head()
```

```python
# aplicamos un offset a toda la columna
stocks.index + pd.DateOffset(days=5)

# tambien podemos retroceder 
stocks.index - pd.DateOffset(days=5)

# adicionanddo meses
stocks.index + pd.DateOffset(months=3)

# retrocediendo un año
stocks.index - pd.DateOffset(years=1)

# horas y minutos
stocks.index + pd.DateOffset(hours=7)

# multiple

stocks.index + pd.DateOffset(years=1, months=3, days=2, hours=14, minutes=23, seconds=12)
```

```python
# freq offset

# encontrar el ibm stock price en cada uno de mis cumpleaños ( april 12, 1991)

# creamos un index con los cumpleaños a buscar
birthdays=pd.date_range( start="1991-04-12", end="2023-04-12", freq=pd.DateOffset(years=1))
# creamos un boolean para filtrar 
stocks.index.isin(birthdays) 
# filtramos el DF
stocks[stocks.index.isin(birthdays)]
```

## Specialized Date offsets
- pandas "anida" date offset mas espcializados en pd.tseries.offsets
- podemos adicionar diferentes cantidades de tiempo a cada fecha, por ej: fin de mes, fin detrimestre, inicio de año.

```python
stocks =pd.read_csv("ibm.csv", parse_dates=["Date"], index_col="Date").sort_index()
stocks.head()
```

```python
stocks.index
```

```python
# podemos redondear cada fecha al mis de mes correspondiente
# sin importar que cada uno requiera un número de días distintos

stocks.index + pd.tseries.offsets.MonthEnd()
    
```

```python
# del mismo modo podemos retroceder al fin del mes anterior
stocks.index - pd.tseries.offsets.MonthEnd()
```

```python
# quarter ends
stocks.index + pd.tseries.offsets.QuarterEnd()
# previus quarter
stocks.index - pd.tseries.offsets.QuarterEnd()

# inicio del trimestre -> parametro: startingMonth
stocks.index + pd.tseries.offsets.QuarterBegin (startingMonth=1)
# inicio del trimestre anterior
stocks.index - pd.tseries.offsets.QuarterBegin (startingMonth=1)

# fin de año 
stocks.index + pd.tseries.offsets.YearEnd()
# inicio de año, al ser suma es el inicio del siguiente año
stocks.index + pd.tseries.offsets.YearBegin()
```

## Timedeltas
- es un objeto de pandas que representa una duración de tiempo
- restando dos objetos Timpestamp se  produce un objeto Timedelta (tambien aplica al restar dos series entre sí)
- el constructor Timedelta acepta parametros de tiempo y strings

```python
# restando timestamps
pd.Timestamp ("2023-03-31") - pd.Timestamp("2023-03-20")
pd.Timestamp ("2023-03-31 12:30:48") - pd.Timestamp("2023-03-20 19:25:59")

# si ponemos el periodo menor primero el resultado seá negativo
pd.Timestamp("2023-03-20 19:25:59") - pd.Timestamp ("2023-03-31 12:30:48")
```

```python
# constructor Timedelta
pd.Timedelta (days=3, hours=2, minutes=5)
pd.Timedelta ("5 minutes")
pd.Timedelta("3 days 2 hours 5 minutes")
```

```python
# Dataset ecommerce (ventas de un ecommerce)
# deseo saber cuanto se demora en entregar la orden
ecommerce = pd.read_csv("ecommerce.csv",index_col="ID", parse_dates=["order_date", "delivery_date"], date_format="%m/%d/%y")
ecommerce.head()
```

```python
# al realizar la resta obtenemos una serie de objetos timedelta
ecommerce["delivery_date"] - ecommerce["order_date"]
```

```python
# almacenamos en el DF
ecommerce ["delivery_time"] = ecommerce["delivery_date"] - ecommerce["order_date"]
ecommerce.head()
```

```python
# si quisiera duplicar el tiempo de entrega realizo:
ecommerce["delivery_date"] + ecommerce["delivery_time"]

# ocurre que el objeto ya es un time delta por lo que puedo realizar operaciones sin problema alguno
```

```python
#almacenamos
ecommerce["dobletiempo"] = ecommerce["delivery_date"] + ecommerce["delivery_time"]
ecommerce.head()
```

```python
# podemos utilizar métodos de pandas
ecommerce["delivery_time"].max()
ecommerce["delivery_time"].min()
ecommerce["delivery_time"].mean()
```
