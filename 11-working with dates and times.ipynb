## Timestamp and DatetimeIndex Objects
- panda posee variadas clases relacionadas a "datetimes"
- el "Timestamp" es similar al objeto "datetime" pero con caracteristicas adicionales
- un DatetimeIndex es un índice de objetos Timestamp
- el constructor Timestamp acepta un string, un objeto datetime ó argumento equi valente a la clase datetime

```python
# tiemstamp: marca de tiempo
pd.Timestamp(2024, 3,27)
pd.Timestamp(2024, 3,27, 16, 11, 5)
# tambien acepta objetos
pd.Timestamp( dt.date(2022, 5,8 ) ) 
pd.Timestamp( dt.datetime(2022, 5,8, 9,55 ) )
# strings
pd.Timestamp("2025-04-13")
pd.Timestamp("2025-04-13  08:35:55")  
```

```python
# tambien se puede representar como serie
pd.Series([pd.Timestamp("2025-04-13  08:35:55")])
# bajo el capot la serie de datetimes contiene timestamps, comprobamos con iloc:
#pd.Series([pd.Timestamp("2025-04-13  08:35:55")]).iloc[0]    
```

## Create range of dates with pd.date_range Function

- la función date_range genera y retona un DatetimeIndex que contiene una secuencia de fechas
- la función necesita al menos 2 de 3 parámetros: start, end y period
- conocidos start y end , pandas asumirá un perido/intervalo diario
- cada elemento de un DatetimeIndex es un Timestamp

```python
pd.date_range(start="2025-01-01", end="2025-01-07")
# es lo mismo que:
pd.date_range(start="2025-01-01", end="2025-01-07", freq="D")

#intervalo  de 2 días
pd.date_range(start="2025-01-01", end="2025-01-07", freq="2D")

# intervalo business days Monday to Friday 
pd.date_range(start="2025-01-01", end="2025-01-07", freq="B") # se excluyen días 4 y 5

# intervalo semanal "W" ( por defecto domingo - EEUU)       
pd.date_range(start="2025-01-01", end="2025-01-31", freq="W")

# intervalo semanal comenzando los viernes
pd.date_range(start="2025-01-01", end="2025-01-31", freq="W-FRI")

# intervalo semanal comenzando los jueves
pd.date_range(start="2025-01-01", end="2025-01-31", freq="W-THU")

# por horas
pd.date_range(start="2025-01-01", end="2025-01-31", freq="H")
pd.date_range(start="2025-01-01", end="2025-01-31", freq="6H")

# por fin de mes
pd.date_range(start="2025-01-01", end="2025-12-31", freq="M")
# por inicio de mes
pd.date_range(start="2025-01-01", end="2025-12-31", freq="MS")

# por fin de año 
pd.date_range(start="2025-01-01", end="2050-12-31", freq="A")
```

```python
# forma alternativa de utilizar esta función

pd.date_range(start= "2012-09-09",freq="D", periods=25 )
pd.date_range(start= "2012-09-09",freq="3D", periods=40 )

pd.date_range(start= "2012-09-09",freq="B", periods=180 )
```

```python
# de la misma forma se puede entregar la fecha final 

pd.date_range(end= "2013-10-31",freq="D", periods=20 )

pd.date_range(end= "2013-12-31",freq="B", periods=75)
```

## the dt attribute

- el atributo dt revela el objeto DatetimeProperties que posee atributos y métodos para trabajar con datetimes, similar al atributo str utilizado en strings.
- el objeto DatetimeProperties pose atributos como: day, month y year, para entregar/manifestar/exponer información acerca de cada fecha en la serie. 
- el método day_name retorna el nombre del día de la semana (string)
- los atributos como  is_month_end y is_quarter_start retorna una serie boleana (útil para filtrar)
