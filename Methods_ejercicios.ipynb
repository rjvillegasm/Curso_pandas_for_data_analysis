```python
def rank_pokemon( pokemon_type):
    if pokemon_type in ["Grass", "Fire" , "Water"]:
        return "Classic"
    elif pokemon_type == "Normal":
        return "Boring"
    else:
        return "TDB"
```

```python
pokemon2.apply(rank_pokemon)

#la funcion se aplicó a cada elemento
```
