```python
import pandas as pd

nba = pd.read_csv("nba.csv")
nba.tail()
```

```python
nba["Name"].sort_values()
nba["Salary"].sort_values()

nba.sort_values(by="Name")
nba.sort_values(by="Name", ascending=True)
nba.sort_values(by="Name", ascending= False)

# por defecto los NaN se van al final
```

```python
nba.sort_values("Age")
nba.sort_values("Age", ascending= True)

nba.sort_values("Age", ascending= False)
```

```python
# el atributo "na_position" nos deja elegir donde ubicar los NaN
nba.sort_values ("Salary", na_position="first" )
nba.sort_values ("Salary", na_position="last" )
# se pueden usar 2 condiciones, dejando los NaN primero y comenzar con el sueldo más alto de forma descendente
nba.sort_values ("Salary", na_position="first", ascending= False )
```

```python
# ordenar por 2 columnas 

# al ordenar por equipo, esta funcionando como si fuera un grup_by
nba.sort_values( by = ["Team", "Name"])


nba.sort_values( by = ["Team", "Name"], ascending=True)

nba.sort_values( by = ["Team", "Name"], ascending= False)
```

```python
# 2 tipos de orden, se usa una lista para guardar los booleanos, de igual cantidad como columnas a ordenar

nba.sort_values( by = ["Team", "Name"], ascending= [True, False])

# en el ejemplo los equipos de forma ascendente y los jugadores descendente
```

```python
nba.sort_values( ["Position", "Salary"], ascending= [False, True])
```

```python
import pandas as pd

nba = pd.read_csv("nba.csv").dropna(how="all")
nba["Salary"]=nba["Salary"].fillna(0).astype("int")
nba.head()
```

```python
# rank crea un ranking, un nuevo valor escalar y lo devuelve junto al index
nba["Salary"].rank()
nba["Salary"].rank(ascending=True)
nba["Salary"].rank(ascending=False).astype("int")

# Creamos una nueva columna , se pudede utilizar mas adelante como parametro de filtrado

nba["Salary_rank"]=nba["Salary"].rank(ascending=False).astype("int")
```
