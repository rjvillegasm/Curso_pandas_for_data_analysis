# Rename index levels

- utiliza el método"set_names" 
- utiliza los parametros "names" y "level" 
- tambien se puede entregar una lista al parámetro "names" para sobreescribir todos los niveles
- "set_names" devuelve una copia, por defecto inplace=False


```python
bigmac.index.set_names(names="Time", level=0)
bigmac.index.set_names(names="Paises", level=1)
# dates no cambia porque la 1ra línea es una copia del DF
# no se ha sobreescrito o utilizado er pafametro inplace
bigmac.index.set_names(names=["Time", "Location"])

#sobreescribir
bigmac.index=bigmac.index.set_names(names=["Time", "Location"])
bigmac.head()
```

# The sort_index method on a multiIndex DF

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], date_format= "%Y-%m-%d", index_col=["Date", "Country"]).sort_index()
bigmac.head()
```

# Extract rows from a multiIndex DF

- a tuple is an inmutable list, can not  be modified after creation
- Create a tuple with a comma between elements, the community convention is to wrao the elements in pharentheses
- the iloc and loc accessors are aviable to extract rows by index position or label
- for the loc accessor , pass a tuple to hold the labels form the index levels 

```python
import pandas as pd

bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], date_format= "%Y-%m-%d", index_col=["Date", "Country"]).sort_index()
bigmac.head()
```

```python
# regular tuple
1,
# other tuple
1, 2
# community convention for readbility
(1, 2)
# comprobamos
type( (1, 2))
```

```python
bigmac.iloc[2]
```

```python
bigmac.loc["2000-04-01"]
```

# importante

- el segundo valor del accessor puede ser el nombre de la columna ó el siguiente valor de la lista (fila)
- se presta para confusiones, ya que al leer el resultado, no se identifica si viene desde la columna o la fila
- se recomienda utilizar la notacion de tule (a, b) (ejemplo3)

```python
# ejemplo 1:
bigmac.loc["2000-04-01", "Canada"]
```

```python
# ejemplo 2:
bigmac.loc["2000-04-01", "Price in US Dollars"]
```

```python
# ejemplo 3:
# utilizando la notacion tuple
bigmac.loc[("2000-04-01", "Canada")]
```

```python
# de este modo el tuple se utiliza para el multiindex 
# permitiendo utilizar el segundo parámetro para la columna
# bigmac.loc[("2000-04-01", "Canada"), "nombre_columna"]

bigmac.loc[("2000-04-01", "Canada"), "Price in US Dollars"] 

# esto nos permitirá extrar slices de data
```

```python
# ejemplo slice:
bigmac.loc[("2000-04-01", "Hungary"):("2000-04-01", "Poland")]
```

```python
# para mejorar  la visibilidad se recomienda usar variables aux

start= ("2000-04-01", "Hungary")
end= ("2000-04-01", "Poland")
bigmac.loc[start:end]
```

```python
# ejemplo end point infinito
bigmac.loc[("2019-07-09", "Hungary"):]
bigmac.loc[:("2019-07-09", "Hungary")]

# bigmac.loc [ (entre esta fila y ésta  ), " de esta columna o columnas"]
```

# The stack method

- moves the column index to the row index
- pandas will return a "multiIndex Series"
- think of it like "stacking" index levels for a multiindex

```python
import pandas as pd

pd.read_csv("worldstats.csv")
```

```python
# como elijo el "outher" para que sea el primer nivel?

pd.read_csv("worldstats.csv").nunique() 

# se concluye que el orden será year-country
```

```python
#pd.read_csv("worldstats.csv", index_col= ["year","country"])
# la cajita que une los años no se vé si no usamos el sort index

world=pd.read_csv("worldstats.csv", index_col= ["year","country"]).sort_index()
world.head()
```

```python
# aplicamos stack 

world.stack()

# de esta forma creamos un nivel adicional en el multiIndex
```

```python
# para regresar a DF 

world.stack().to_frame()

# la columna queda sin nombre 
```

# the unstack method

- moves a row index to the column index
- por default mueve índice " mas interno"
- utiliza el parámetro "level"
- "level" acepta posición ó nombre ó lista de posición/nombres 

```python
world=pd.read_csv("worldstats.csv", index_col= ["year","country"]).sort_index().stack()
world.head()
```

```python
world.unstack()
```

```python
# para revisar usamos .columns

world.unstack().columns

#otro ejemplo:
#world.unstack().unstack().columns
```

```python
# custom level

world.unstack(level=0)

world.unstack(level="year")
```

# the pivot method

- the pivot method reshapes data for  a tall fortmat to a wide format
- el parámetro "index" setea el índice horizontal del DF pivoteado
- el parámetro "columns" setea la columna, cuyos valores serán las nuevas columnas del DF pivoteado
- el parámetro "values" setea los valores del DF pivoteado, pandas populará los valores correctos basado en la intersección de índice y columnas  

```python
import pandas as pd
sales =pd.read_csv("salesmen.csv")

sales
```

```python
# entregamos los parámetros deseados
sales.pivot(index="Date", columns="Salesman", values="Revenue")
```

# the melt method

- es el inverso al pivot
- convierte una tabla ancha a una alta
- es el metodo ideal cuando tienes multiples columnas guardando el mismo " data point"
- "ask yourself wheter the column´s values ara a type of the column header, if they are not , the data is likely stored in a wide format"
- Pregúntese si los valores de la columna son un tipo de encabezado de columna; si no lo son, es probable que los datos estén almacenados en un formato amplio.
- el parámetro "id_vars" acepta la columan cuyos valores serán repetidos para cada columna ( la columna que no cambia)
- el parámetro "var_name" setea el nombre de la nueva columna donde agrupo los antiguos encabezados, es decir,lo que representan aquellos headers
- el parámetro "value_name" setea el nuevo nombre de la columna, donde se aguparan los valores que estaban en aquellas columnas



```python
quarters  = pd.read_csv("quarters.csv")
quarters.head()
```

```python
quarters.melt(id_vars="Salesman", var_name="Quarter", value_name= "Revenue")
```

```python
# tambien se puede utlizar solo entregando id_vars
quarters.melt(id_vars= "Salesman")
```

# the pivot_table method

- funciona similar a las tablas pivote de excel
- la tabla pivote es una tabla cuyos valores son agregaciones de grupos de valores desde otra tabla
- el parámetro "values" acepta la columna numérica donde serán agrupados los valores
- el parámetro "aggfunc" declara la funcion de agregación ( por defecto mean/average)
- el parámetro " index" setea los índices de la tabla ( permite multiIndice)
- el parámetro "columns" setea las etiquetas de las columnas ( permite multiIndice)

```python
foods = pd.read_csv("foods.csv")
foods.head()
```

```python
# elijo la columna numérica a resumir/aggregated 
# elijo que categorías quiero comparar : item, ciudades ó géneros
foods.pivot_table( values="Spend", index="Gender" )
foods.pivot_table( values="Spend", index="Gender", aggfunc="mean" )
foods.pivot_table( values="Spend", index="Gender", aggfunc="sum" )


#foods.pivot_table( values="Spend", index="City" )
#foods.pivot_table( values="Spend", index="Item" )
```

```python
# ejemplo multiindex

foods.pivot_table( values="Spend", index=["Gender", "City"], aggfunc="sum" )
#foods.pivot_table( values="Spend", index=["Gender", "Item"], aggfunc="sum" )
```

```python
# parámetro columns, con multiindex
foods.pivot_table( values="Spend", index=["Gender", "Item"],columns="City", aggfunc="sum" )
```

```python
# parámetro columns multiindex
foods.pivot_table( values="Spend", index="Item",columns=["Gender", "City"], aggfunc="sum" )
```

```python
# agg funcionts
# sum - mean - count - max - min 

foods.pivot_table( values="Spend", index="Item",columns=["Gender","City"], aggfunc="count" )
```
