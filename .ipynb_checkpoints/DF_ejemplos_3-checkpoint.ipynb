# Difference Between loc and iloc

## The difference between the loc and iloc functions is that the loc function selects rows using row labels (e.g. tea) whereas the iloc function selects rows using their integer positions (staring from 0 and going up by one for each row).

```python
bond.iloc[5]

# Entrega la información de una fila seleccionada en el paréntesis
```

```python
# se le puede entregar una lista , para consultar múltiples filas
# cuidado con el paréntesis cuadrado
# el resultado es un pequeño dataframe

bond.iloc[[15,20]]
# aplican las reglas de series 
# bond.iloc[4:8]
# bond.iloc[0:6]
#bond.iloc[:6]
bond.iloc[20:]
```

 Second arguments to loc and Iloc accesors

```python
# usamos "film" como index
# ordenar el índice  da velocidad a pandas
bond = pd.read_csv("jamesbond.csv", index_col= "Film").sort_index()
bond.head()
```

```python
# al entregar un segundo argumento, le damos la posicion exacta en una matriz

bond.loc["Diamonds Are Forever", "Director"]

# el argumento tambien puede ser una lista, ej:
bond.loc[["Octopussy", "GoldenEye"], "Director"]

# tambien es posible obtener slices, debemos conocer el orden de las columnas

bond.loc[["Octopussy", "GoldenEye"], "Director":"Budget"]

# dos slices

bond.loc[ "GoldenEye":"Octopussy", "Director":"Budget"]

# entregamos una lista con las columnas específicas, en orden específico, similar a una query
bond.loc[ "GoldenEye":"Octopussy", ["Actor" , "Bond Actor Salary", "Year"]]

```

```python
# De igual forma que con loc podemos hacerlo con iloc
# forma similar al algebra de matrices

bond.iloc[0,2]
bond.iloc[3,5]
9
# se puden ulizar listas , slices como en la funcion loc
```

# precaución

```python
#obtenemos un booleano
bond["Actor"] == "Sean Connery"
# obtenemos el subset de films protagonizados por S.C.
bond[bond["Actor"] == "Sean Connery"]
# reemplazamos utilizando Loc accesor
bond[bond["Actor"] == "Sean Connery"].loc[:, "Actor"] = " Sir Sean Connery"

# al realizar este reemplazo en un subset o view, este no se ha guardado en nuestro DF original
# bond.head(15)
```

```python
# la forma correcta es utilizando una variable
# la variable almacena la serie booleana
is_sean_connery = bond["Actor"] == "Sean Connery"
# filtramos
bond.loc[is_sean_connery]
# localizamos (individualizamos/ target) la columna Actor y reemplazamos todos los valores por "Sir Sean Connery"
bond.loc[is_sean_connery , "Actor"] = "Sir Sean Connery"
#comprobamos
bond
```

The "apply" method with DF
- aplica una funcion ó serie de pasos a cada fila

```python
bond["Actor"].apply(len)
```

```python
# ejemplo  Movie Ranking System
#
# Condition -> Designation
# 80s movie -> "Great 80's flick"
# Sean Connery -> "The best Bond ever"
# Budget >100  -> "Expensive movie, fun"
# Others -> "No Comment"
```

```python
#primer paso
def rank_movie(row):
    print(row)
    
bond.apply(rank_movie, axis="columns")

# se obtine un objeto donde se recorre la fila
# avanzando una columna a la vez
# retorna el valor de cada columna
# simil a transponer en excel
```

```python
# axis contrario 
def rank_movie(row):
    print(row)
    
bond.apply(rank_movie, axis="index")

# se obtine un objeto donde se recorre la columna
# avanzando una fila a la vez
# retorna el valor de cada fila
# simil a transponer en excel
```

```python
# segundo paso
# usamos variables locales para extraer las columnas objetivo
# aplicamos los condicionales

def rank_movie(row):
    year=row.loc["Year"]
    actor=row.loc["Actor"]
    budget=row.loc["Budget"]
    
    if year >=1980 and year < 1990:
        return "Great 80's flick!"
    
    if actor == "Sean Connery":
        return "The Best Bond Ever"
    
    if budget >100:
        return "Expensive movie, fun"
    
    return " No comment"
    
    
bond.apply(rank_movie, axis="columns")
```
