# the groupby object


- éste método retorna un objeto DataFrameGroupby
- el objecto DataFrameGroupby puede realizar operaciones de aggregación en cada grupo

```python
fortune =pd.read_csv("fortune1000.csv")
fortune.head()
```

```python
# sum of revenue for each sector

fortune.groupby("Sector")

#asignamos
sectors = fortune.groupby("Sector")
sectors
# se ha creado un DF para cada sector existente en la columna
```

```python
len(sectors)
# entrega la cantidad de sectores
```

```python
sectors.size()
#entrega la cantidad de empresas en cada sector
```

```python
sectors.first()
# entrega cada uno de los sectores y solo una empresa (la primera)
```

```python
sectors.last()
# entrega cada uno de los sectores y solo una empresa (la última)
```

# Retrive a group with the get_group method

- el método get_group en el objeto DFGBy entrega uno de los DF anidados perteneciente a un grupo/categoría específica 


```python
#ejemplos:
sectors.get_group("Technology")
sectors.get_group("Energy")
```

# Methods on the Group by object

- se utilizan brackets en el objeto DFGBy para extraer una columna del DF original
- el obejto SeriesGroupBy tendrá metodos de agregación disponibles
-  pandas ejecutará el cálculo en cada uno de los grupos en su colección
- por ejemplo el método "sum" agupará el total de "revenue" para cada grupo/categoría


```python
sectors["Revenue"]

# entrega un objeto SeriesGroupBy
```

```python
sectors["Revenue"].sum()
```

```python
# equivale a filtrar cada sector y luego ralizar la aggregación ( 21 veces)
mask = fortune["Sector"]== "Apparel"
fortune [mask] ["Revenue"].sum()
```
