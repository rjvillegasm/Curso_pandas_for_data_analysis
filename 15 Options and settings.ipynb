### Changing option swith attributes

- el módulo <code>pd.options </code>configura las opciones de pandas
- el módulo <code> display</code> para configurar la presentación de los datos
- el atributo <code>min_rows</code> setea el mínimo de de filas cuando se excede <code>max_rows</code>
- el atributo <code> max_columns</code> setea el número maximo de columnas, no existe un atributo "min_columns"
- para sobrescribir un atributo se tutliza el signo igual  

```python
# generando números aleatoreos

np.random.randint(0,100, [60,50])

# creamos un DF y asignamos

df=pd.DataFrame(np.random.randint(0,100, [61,50]) )
df
```
