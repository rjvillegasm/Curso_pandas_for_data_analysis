# Grouping by multiple columns

- se entrega una lista de columnas al método groupby para agrupar por pares de valores a través de las columnas
- targetea una columna a extraer/recuperar el objeto SeriesGroupby, luego raliza la agregación
- pandas retornará una serie multi índice, donde los niveles serán los grupos originales


```python
import pandas as pd
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
```

```python
fortune.head()
```

```python
# combinamos sector+industria

sectors=fortune.groupby(["Sector", "Industry"])

#revisamos
sectors.size()
# de este modo obtenemos 79 combinaciones únicas entre sector e industria
```

```python
sectors["Revenue"].sum()
# de este modo obtenos la suma de revenue en cada sector y por industrias (si huviese mas de una)

# sectors["Employees"].mean().head()
```

# The agg method

- el método agg aplica distintos métodos de agg a diferentes columnas
- se puede invocar directamente en el objeto DFGby
- se entrega un diccionario donde las keys son las columnas y los valores representan a os metodos de aggregación
- se obtine un DF

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")
```

```python
# simple
sectors.agg({"Revenue":"sum"})

#multiple 
sectors.agg({"Revenue":"sum", "Profits":"max", "Employees":"mean"})
```

# iterando a travez de los grupos

- el objeto DFGBy seoporta el método apply ( al igual que las series y DF)
- éste método invoca una función en cada DF anidado en el objeto DFGby
- entrega un DF con el valor resultante de las funciones

```python
fortune = pd.read_csv("fortune1000.csv", index_col="Rank")
sectors = fortune.groupby("Sector")

fortune.head()
```

```python
# find the two companies in each sector with the most companies

# metodo tradicional  en DF
#fortune.nlargest(2, "Employees")

#creamos una funcion
def top_two_companies_by_employee_count(sector):
    return sector.nlargest(2,"Employees")
#usamos apply para que el entrege los 21 grupos/sectores 
sectors.apply(top_two_companies_by_employee_count)
```
